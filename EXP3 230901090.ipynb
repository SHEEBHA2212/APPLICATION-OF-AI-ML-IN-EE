```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("MOTORPRICE.xlsx")

print(data.head())
```

```python
X = data[['year']]   # Independent variable
y = data['cost']    # Dependent variable
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
future_year = np.array([[2026]])
predicted_price = model.predict(future_year)

print("Predicted motor price in 2026:", predicted_price[0])
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Motor Price")
plt.title("Motor Price Prediction using Linear Regression")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
data = pd.read_excel("MOTORPRICE.xlsx")

# Independent and dependent variables
X = data[['year']]      # Year
y = data['cost']        # Motor price

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict values for existing data
y_pred = model.predict(X)

# Calculate R² score
r2 = r2_score(y, y_pred)
print("R² Score:", r2)

# Predict motor price for year 2026
year_2026 = pd.DataFrame([[2026]], columns=['year'])
price_2026 = model.predict(year_2026)

print("Predicted motor price for 2026:", price_2026[0])

# Plot the graph
plt.figure()
plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, label="Regression Line")
plt.scatter(year_2026, price_2026, label="Prediction for 2026")
plt.xlabel("Year")
plt.ylabel("Motor Price")
plt.title("Linear Regression: Motor Price Prediction")
plt.legend()
plt.show()
```

```python
m = model.coef_[0]      # slope
c = model.intercept_   # intercept

print("Slope (m):", m)
print("Intercept (c):", c)
```
